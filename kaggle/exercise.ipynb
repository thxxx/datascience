```python
tdf = pd.read_csv('train.csv')
tdf
```

```python
t_bool = tdf[tdf['Age']>65]
#나이가 60세 이상이고, 선실등급이 1등급이며, 성별이 여성인 승객 찾기.

tdf[ ( tdf['Age']>60 ) & (tdf['Pclass']==1) & (tdf['Sex']=='female') ]

mtsixty = tdf['Age']>60
pcone = tdf['Pclass']==1
femalet = tdf['Sex']=='female'

print(tdf[mtsixty & pcone & femalet])
```

```python
tdf.sort_values(by=['Name'])
print(tdf.iloc[0:3, 0:6].head(3))

tdf.sort_values(by=['Name'], inplace=True)
print(tdf.iloc[0:3, 0:6].head(3))

tdf.sort_values(by=['Name'], inplace=True, ascending=False)
print(tdf.head(3))
```

```python
tdfg = tdf.groupby(by='Pclass')
tdfg.agg({ 'Age':'max', 'SibSp':'sum', 'Fare':'mean' })
```

```python
tdf['Cabin'] = tdf['Cabin'].fillna('C000')
tdf['Age'].fillna((tdf['Age'].mean()), inplace=True)
tdf['Embarked'].fillna(('s'), inplace=True)
tdf.isna().sum()
```

```python
tdf['Name_len'] = tdf['Name'].apply(lambda x: len(x))
print(tdf[['Name', 'Name_len']].head(3))
```

```python
tdf['ChildAdult'] = tdf['Age'].apply(lambda x : 'Child'if x<=18 else 'Adult')
print(tdf[['Age', 'ChildAdult']].head(10))
```
